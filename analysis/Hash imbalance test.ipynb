```python
import random
import matplotlib.pyplot as plt
import statistics as st
x = [[] for i in range(32)]
```

```python
A=[]
for i in range(100):
    a=random.randint(1,1000)
    A.append(a)
```

```python
for each in A:
    index=each%32
    x[index].append(each) 
```

```python
i=0
counter=[]
for each in x:
    counter.insert(i,len(each))
    #print(len(each))
    i+=1
```

```python
plt.plot(range(32),counter)
```

```python
st.mean(counter)
```

```python
st.stdev(counter)
```
